```python
#Project #1
#Danceability charts/graphs

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import os
import json

#Load the .csv file
loadFile = "nueTopTens.csv"

#Create a dataframe
songList = pd.read_csv(loadFile,encoding = "ISO-8859-1")
songList.head()
```

```python
dnce_df = songList[["top genre", "simplified genres"]]

#dnce_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['dnce']
dnce_combo = dnce_df.nunique()

#line_x = dnce_agg.index.values.tolist()
#line_y = dnce_agg["mean"].tolist()

dnce_combo.plot(kind= 'bar', figsize=(10,4))
#plt.grid()
plt.xlabel("")
plt.ylabel("Unique Genres")
plt.title("Number of Genres Before and After Data Cleanup")
plt.show()
```

```python
dnce_df = songList[["year", "dnce", "nrgy", "simplified genres"]]

dnce_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['dnce']

line_x = dnce_agg.index.values.tolist()
line_y = dnce_agg["mean"].tolist()

plt.plot(line_x, line_y, marker='x')
plt.grid()
plt.xticks(line_x)
plt.xlabel("Year")
plt.ylabel("Average Danceability")
plt.title("Average Danceability by Year")
plt.show()
```

```python
dnce_df = songList[["year", "dnce", "nrgy", "simplified genres"]]

dnce_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['dnce']
nrgy_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['nrgy']

line_x = dnce_agg.index.values.tolist()
line_y = dnce_agg["mean"].tolist()
line2_x = nrgy_agg.index.values.tolist()
line2_y = nrgy_agg["mean"].tolist()

plt.plot(line_x, line_y, label = "dance")
plt.plot(line2_x, line2_y, label = "energy")
plt.grid()
plt.xticks(line_x)
plt.xlabel("Year")
plt.ylabel("Average")
plt.title("Average Danceability and Energy per Year")
plt.legend()
plt.show()
```

```python

dnce_df = songList[["year", "bpm", "nrgy", "dnce", "val", "dur", "acous", "simplified genres"]]

bpm_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['bpm']
nrgy_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['nrgy']
dnce_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['dnce']
val_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['val']
dur_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['dur']
accous_agg = songList.groupby('year').agg(['mean', 'median', 'var', 'std', 'sem'])['acous']

line_x = bpm_agg.index.values.tolist()
line_y = bpm_agg["mean"].tolist()
line1_x = nrgy_agg.index.values.tolist()
line1_y = nrgy_agg["mean"].tolist()
line2_x = dnce_agg.index.values.tolist()
line2_y = dnce_agg["mean"].tolist()
line3_x = val_agg.index.values.tolist()
line3_y = val_agg["mean"].tolist()
line4_x = dur_agg.index.values.tolist()
line4_y = dur_agg["mean"].tolist()
line5_x = accous_agg.index.values.tolist()
line5_y = accous_agg["mean"].tolist()

plt.plot(line_x, line_y, label = "bpm")
plt.plot(line1_x, line1_y, label = "energy")
plt.plot(line2_x, line2_y, label = "dance")
plt.plot(line3_x, line3_y, label = "valence")
plt.plot(line4_x, line4_y, label = "duration")
plt.plot(line5_x, line5_y, label = "accoustic")
plt.grid()
plt.xticks(line_x)
plt.xlabel("Year")
plt.ylabel("Average")
plt.title("Summary by Year")
plt.legend()
plt.rcParams["figure.figsize"] = (10,5)
plt.show()
```
