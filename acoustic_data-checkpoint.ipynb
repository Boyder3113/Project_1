```python
#Dependancies 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
from sklearn import datasets
import scipy.stats as sts
import sklearn.datasets as dta
import scipy.stats as st
import seaborn as sns
```

```python
#Import data and verify
source_data = pd.read_csv('nueTopTens.csv', encoding = 'ISO-8859-1')
source_data.head()
```

```python
# ID data types
source_data.dtypes
```

```python
# Key data for study.
acoustic_df = pd.DataFrame(source_data, columns = ['year', 'simplified genres', 'acous', 'pop'])
acoustic_df
```

```python
#Create DataFrame for acoustic data
acous_pop_year = acoustic_df.groupby('year').acous.agg(['count','mean']).rename(columns = {'mean':'Avg Acoustic Rating','count':'#1 Songs'})
acous_pop_year
```

```python
# Line graph to show average acoustic rating of songs over time.
raw_chart = acous_pop_year.plot(kind = 'line', title = 'Song Count to Avg Acoustic Rating', figsize = (10,5))
raw_chart.set_ylabel('Count/Rating')
raw_chart.set_xlabel('Year')
plt.grid()
```

```python
#Area Chart Unstacked
# Merged from enich_analytics
x = acous_pop_year.index.values.tolist()
y1 = acous_pop_year['#1 Songs'].values.tolist()
y2 = acous_pop_year['Avg Acoustic Rating'].values.tolist()
mycolors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink', 'tab:olive']      
columns = ['#1 Songs', 'Avg Acoustic Rating']

# Draw Plot 
fig, ax = plt.subplots(1, 1, figsize=(12,8), dpi= 80)
ax.fill_between(x, y1=y1, y2=0, label=columns[0], alpha=0.9, color=mycolors[3], linewidth=2)
ax.fill_between(x, y1=y2, y2=0, label=columns[1], alpha=1.0, color=mycolors[1], linewidth=2)

# Decorations
ax.set_title('Count of #1 Songs vs Avg Acoustic Rating', fontsize=18)
ax.set(ylim=[0, 30])
ax.legend(loc='best', fontsize=12)
plt.xticks(x[::1], fontsize=10, horizontalalignment='center')
plt.yticks(np.arange(10, 105.0, 10), fontsize=10)
#plt.xlim(-10, x[-1])
plt.xlim(x[0], x[-1])

# Draw Tick lines  
for y in np.arange(2.5, 30.0, 2.5):    
    plt.hlines(y, xmin=0, xmax=len(x), colors='black', alpha=0.3, linestyles="--", lw=0.5)
   
# Lighten borders
plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.3)
plt.savefig('output_data/count_pop')
plt.show()
```

```python
# Scotter plot with r squared value
x_values = acoustic_df['pop']
y_values = acoustic_df['acous']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Acousticness')
plt.ylabel('Popularity')
print(f'The r-squared is {rvalue**2}')
plt.show()
```

```python
# Quartile Calculations
acoustic = acoustic_df['acous']
quartiles = acoustic.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of Acousticness is: {lowerq}")
print(f"The upper quartile of Acousticness is: {upperq}")
print(f"The interquartile range of Acousticness is: {iqr}")
print(f"The the median of Acousticness is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```
